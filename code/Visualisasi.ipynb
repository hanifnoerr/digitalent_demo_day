```python
#load library
import pandas as pd
import numpy as np
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('tweets_final.csv') #load filenya
df.head(3)
```

```python
df['keyword'].nunique() #cari tau ada berapa keyword
```

```python
print(df['keyword'].unique()) #keyword yg tersedia
```

```python
#pilih topicnya
key = 'fatalities'
df_topic=df[df['keyword'] == key]
```

```python
df_topic
```

```python
text = " ".join(i for i in df_topic.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
colors = np.array(["gray", 'orangered', 'hotpink', 'gold', 'teal', 'cornflowerblue'])

plt.bar(df_topic.Sentiment.value_counts().index,df_topic.Sentiment.value_counts().values, color=colors)
plt.title('Sentimen terkait kata kunci '+key)
plt.show()
```
